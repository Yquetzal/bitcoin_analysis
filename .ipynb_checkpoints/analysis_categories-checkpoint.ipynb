```python
def write_formated_plot(fig,location):
    fig.layout.update(showlegend=False)
    fig.update_layout(
    margin=dict(l=10, r=10, t=5, b=5))
    fig.update_traces(mode="lines", hovertemplate=None)
    fig.update_layout(hovermode="x unified")
    pio.write_html(fig, file=location, include_mathjax=False,include_plotlyjs=False,full_html=False)
    return fig
```

```python
def get_monthly_stat(df):
    agg = df.groupBy("year","month").agg(countDistinct("hash").alias("nb_transactions"),
                                         countDistinct(when(col("src")=="mining",col("hash"))).alias("nb_mining"),
                                   (sum("value")/100000000).alias("sum_btc"),
                                   count("value").alias("nb_outputs"),
                                   sum("valueUSD").alias("sum_USD"),
                                    (sum(when(col("src")=="mining",col("value")))/100000000).alias("mining_btc"),
                                    (sum(when(col("src")=="mining",col("valueUSD")))).alias("mining_USD"),

                                    count(when(col("src")=="mining","count")).alias("nb_mining_outputs"),
                                   mean("PriceUSD").alias("PriceUSD"),
                                   count(when(col("spent_timestamp").isNull(),"value")).alias("nb_unspent"),
                                   mean((col("spent_timestamp")-col("time"))/60/60).alias("spending_time"))
    to_return= agg.select("*").toPandas()
    to_return = to_return.set_index(pd.to_datetime(to_return['year'].astype(str)  + to_return['month'].astype(str), format='%Y%m'))
    to_return = to_return.sort_index()
    return to_return
```

```python
monthly = get_monthly_stat(data)
```

```python
def get_tr_stats(df):
    agg = df.groupBy("hash").agg(sum("value").alias("tr_value"),
                                 sum("valueUSD").alias("tr_valueUSD"),
                                 mean("nb_inputs").alias("nb_inputs"),
                                 mean("nb_outputs").alias("nb_outputs"),
                                 first("src").alias("src"),
                                 first("src_identity").alias("src_identity"),
                                 first("src_cat").alias("src_cat"),
                                 mean("fee").alias("fee"),
                                 first("year").alias("year"),
                                 first("month").alias("month"),
                                 first("PriceUSD").alias("PriceUSD")
                                )\
    .withColumn("fee_for100",(col("fee")/(col("tr_value")+col("fee")+1))*100)\
    .withColumn("feeUSD",col("fee")/100000000*col("PriceUSD"))\
    .groupBy("year","month").agg((sum("fee")/100000000).alias("sum_fee_btc"),
                                 sum("tr_value").alias("tr_value"),
                                 mean("fee").alias("mean_fee_satoshi"),
                                 mean("feeUSD").alias("mean_feeUSD"),
                                 mean("fee_for100").alias("mean_fee_for100"),
                                 count("fee").alias("nb_transactions"),
                                 mean("nb_inputs").alias("mean_nb_inputs"),
                                 (sum(when(col("src")=="mining",col("value")))/100000000).alias("mean_tr_mining_btc"),
                                 (sum(when(col("src")=="mining",col("tr_valueUSD")))/100000000).alias("mean_tr_mining_USD"),
                                 mean("nb_outputs").alias("mean_nb_outputs"))
    agg = agg.withColumn("mean_fee_for100_overall",(col("sum_fee_btc")/((col("tr_value")+1)/100000000+col("sum_fee_btc")))*100)
    to_return= agg.select("*").toPandas()
    to_return = to_return.set_index(pd.to_datetime(to_return['year'].astype(str)  + to_return['month'].astype(str), format='%Y%m'))
    to_return = to_return.sort_index()
    return to_return
```

```python
monthly_tr = get_tr_stats(data.filter(col("year")==2010))
monthly_tr = get_tr_stats(data)
```

```python
def get_mining_stats(df):
    agg = df.filter(col("src")=="mining").groupBy("hash").agg(sum("value").alias("value"),
                                 sum("valueUSD").alias("valueUSD"),
                                 first("nb_outputs").alias("nb_outputs"),
                                first("year").alias("year"),
                                 first("month").alias("month"),
                            first("time").alias("time"),
                                mean("spent_timestamp").alias("spent_timestamp")
    )\
    .groupBy("year","month").agg(count("value").alias("nb_mining"),
                                 (sum("value")/100000000).alias("sum_mining_btc"),
                                 sum("valueUSD").alias("sum_mining_USD"),
                                 (mean("value")/100000000).alias("mean_mining_output_btc"),
                                 mean("valueUSD").alias("mean_mining_output_USD"),
                                mean("nb_outputs").alias("mean_nb_outputs"),
                                mean(when(col("value")>0,col("nb_outputs"))).alias("mean_nb_outputs>0"),
                                 mean((col("spent_timestamp")-col("time"))/60/60).alias("mining_spending_time")
)
    to_return= agg.select("*").toPandas()
    to_return = to_return.set_index(pd.to_datetime(to_return['year'].astype(str)  + to_return['month'].astype(str), format='%Y%m'))
    to_return = to_return.sort_index()
    return to_return
```

```python
def plot_cumul(df,category,value):
    pd_df= df.select("*").toPandas()
    pd_df = pd_df.set_index(pd.to_datetime(pd_df['year'].astype(str)  + pd_df['month'].astype(str), format='%Y%m'))
    pd_df = pd_df[[category,value]].pivot(columns=category)
    pd_df.plot.area(figsize=(10,7))
```

```python
group_by_scale = with_size_class.groupBy("year","month","scale").agg(sum("count_tr").alias("total_tr"))
plot_cumul(group_by_scale,"scale","total_tr")
```

```python
group_by_scale = with_size_class.groupBy("year","month","scale").agg(sum("sum_satoshi").alias("satoshi"))
plot_cumul(group_by_scale,"scale","satoshi")
```
